```python
import pandas as pd
df=pd.read_csv("/home/mansi/Desktop/internship/cognoRise_internship_task/task2/data.csv")
df
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()

df["city"]=l.fit_transform(df["city"])
df["city"]
```

```python
x=(df[["sqft_living","bedrooms","city"]])
x
```

```python
y=df.iloc[:,1:2]
y

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)

print("length of x_train data is ",len(x_train))
print("length of y_train data is ",len(y_train))
```

```python
x_train
```

```python
y_train
```

```python
#1.linear Regression
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l

model=l.fit(x_train,y_train)
model

pred=l.predict(x_test)
pred

```

```python
from sklearn.metrics import r2_score,mean_squared_error
print(r2_score(y_test,pred))
print(mean_squared_error(y_test,pred))
```

```python
input=[[1700,4,33]]
prediction=model.predict(input)
prediction
```

```python
#plot the result
import matplotlib.pyplot as plt
plt.scatter(y_test,pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```

```python
#2.random forest regression
from sklearn.ensemble import RandomForestRegressor
r=RandomForestRegressor(max_depth=4)

model=r.fit(x_train,y_train)
model

pred=model.predict(x_test)
pred
```

```python
print("r2 score ",r2_score(y_test,pred))
print("mean_squared error ",mean_squared_error(y_test,pred))
```

```python
#plot the result
import matplotlib.pyplot as plt
plt.scatter(y_test,pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```

```python
#3.decision tree regression
from sklearn.tree import DecisionTreeRegressor
d=DecisionTreeRegressor(max_depth=3)
d

model=d.fit(x_train,y_train)
model

pred=model.predict(x_test)
pred
```

```python
print("r2 score ",r2_score(y_test,pred))
print("mean_squared error ",mean_squared_error(y_test,pred))
```

```python
#4.gradient boosting regression
from sklearn.ensemble import GradientBoostingRegressor
g=GradientBoostingRegressor( max_depth=2, learning_rate=0.1)
g

model=g.fit(x_train,y_train)
model

pred=model.predict(x_test)
pred
```

```python
print("r2 score ",r2_score(y_test,pred))
print("mean_squared error ",mean_squared_error(y_test,pred))
```

```python
#plot the result
import matplotlib.pyplot as plt
plt.scatter(y_test,pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```
