```python
#read csv file first
import pandas as pd
df=pd.read_csv("/home/mansi/Desktop/internship/cognoRise_internship_task/data.csv")
df
```

```python
#data preprocessing 

df.isnull().sum()
df.dropna(inplace=True)  #drop null values and rows first
 
```

```python
#identify independent var , and depdendent var
x=df.iloc[:,2:].values #except diagnosis all columns as a independent var
# print(x)


y=df.iloc[:,1:2].values #diagnosis = depdendent var
# print(y)


# divided data into train and test data using holdout method
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3) #0.2 means 20 % data will be used for testing purpose

print("length of x training data set is ",x_train.shape)
print("length of y training data set is ",y_train.shape)

print("length of x testing data set is ",x_test.shape)
print("length of y testing data set is ",y_test.shape)
```

```python

print("x train data is \n ",x_train)
print("y train data is \n ",y_train)
```

```python

print("x test data is ",x_test)
print("y test data is ",y_test)
```

```python
#algorithms

#1.KNN

from sklearn.neighbors import KNeighborsClassifier
k=KNeighborsClassifier()

model=k.fit(x_train,y_train) #train model using training data
pred=model.predict(x_test) #prediction based on testing data
print("predicted values is ",pred)
```

```python
#check accuracy of Algorithm

from sklearn.metrics import accuracy_score,classification_report
print("accuracy score of KNN Algorithm is ",accuracy_score(y_test,pred))
print("classification report of KNN Algorithm : \n \n",classification_report(y_test,pred))
```

```python
#plotting those predicted values so that it can easily be understandable

#confusion matrix
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
c=confusion_matrix(y_test,pred)
print("Confusion matrix is \n",c)

cmat=ConfusionMatrixDisplay(confusion_matrix=c)
cmat.plot()
```

```python
#2.Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
r=RandomForestClassifier()
r

#need to convert data into nuemric category
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
y=l.fit_transform(y)

#after converting numeric category again need to divide data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)

model=r.fit(x_train,y_train)
model

pred=model.predict(x_test)
print("prediction is : \n ",pred)

```

```python

print("accuracy score is ",accuracy_score(y_test,pred))
print("classfication-report  is : \n \n ",classification_report(y_test,pred))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
c=confusion_matrix(y_test,pred)
print("Confusion matrix is \n",c)

cmat=ConfusionMatrixDisplay(confusion_matrix=c)
cmat.plot()
```

```python
#3.linear Regression 
#note that : linear regression cannot be used for binary classification if u try to use it will give you wrong predictions....

from sklearn.linear_model import LinearRegression
l=LinearRegression()
l

model=l.fit(x_train,y_train)
model

pred=model.predict(x_test)
print("Prediction \n : ",pred)


from sklearn.metrics import r2_score,mean_squared_error
print("R2 Score is ",r2_score(y_test,pred))
print("Mean Squared Error is ",mean_squared_error(y_test,pred))

import matplotlib.pyplot as plt
plt.scatter(y_test,pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.show()
```
